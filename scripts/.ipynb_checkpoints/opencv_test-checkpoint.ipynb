```python
import cv2
from jetbot import Camera
```

```python
help(Camera)
```

```python

cap = cv2.VideoCapture(0)
```

```python
cap.set(cv2.CAP_PROP_FRAME_WIDTH,640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,480)
```

```python
while cv2.waitKey(33)<0:
    ret, frame = cap.read()
    cv2.imshow("VideoFrame",frame)

cap.release()
cv2.destroyAllWindows()
```
